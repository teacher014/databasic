**동전 앞면/뒷면 발생 비율 계산하기**

* 함수 : 없음
* 입력 : 시뮬레이션 횟수 : 1~10000
* 출력 : 입력한 횟수까지 동전의 앞면 발생 비율
* 변수 : 
  * n : 시뮬레이션 횟수
  * coinF : 동전의 앞면 발생 횟수
  * cFR : 난수에 의한 동전의 앞면(0)/뒷면(1)



````python
import random

n = int(input("simul count:"))
if n < 1 or n > 10000:
  print('Simul count = 1 ~ 10000')

else:
  print("count : Ratio")
  coinF = 0
  for i in range(1, n+1):
    cFR = random.randint(0,1)
    if cFR == 0: coinF += 1
    print(f'{i:^5} {(coinF / i * 100):^10.2f}') # <10 왼쪽정렬, ^10 가운데정렬 >10 오른쪽정렬
````

**동전앞면나올 확률 구하기**<br>

0 : 동전의 뒷면
1 : 동전의 앞면

순서
1. 0 or 1 랜덤으로 뽑아내기
2. 뒷면이 나올 확률 : 0 나온 횟수 / 전체 시행 횟수<br>
   앞면이 나올 확률 : 전체 시행 횟수 - 0 나온 횟수 / 전체 시행 횟수 
3. 시각화

````python
import random

cnt = [0,0]
for _ in range(100):
  cnt[0] += random.randint(0,1)
cnt[1] = 100-cnt[0]
print(cnt)
````

````python
import random
import matplotlib.pyplot as plt

cnt = [0,0]
for _ in range(1000):
  cnt[0] += random.randint(0,1)
cnt[1] = 1000-cnt[0]

print(cnt)
plt.pie(cnt, labels=['Tails0','Heads1'], autopct='%1.2f%%')
````

**동전을 10번 던져서 앞면이 나오는 횟수 세보기를
10만번 반복하기** 

````python
import random
num=[random.randint(0,1) for _ in range(10)]
num
#앞면이 나오는 횟수 num 리스트 요소(1)의 합
sum(num)
````

````python
import random
import matplotlib.pyplot as plt

res = [0]*11
for i in range(100000):
  cnt = sum([random.randint(0,1) for _ in range(10)])
  res[cnt] += 1

for i in range(11):
  print(f'앞면이 {i}번 나올 확률 : {res[i]/1000000} {res}')
````

````python
plt.bar(range(11), res, tick_label=range(11))
plt.title('coin prob')
plt.ylabel('prob')
plt.xlabel('count')
````

**주사위 확률**

````python
import random

print('주사위를 던집니다!')
print('결과: ', random.randint(1,6))

total = int(input('횟수:'))

dice = [0]*6

for i in range(total):
    num = random.randint(1,6)
    dice[num-1] += 1

for idx, i in enumerate(dice):
    print(f'{total}번 중 {idx+1}이 나온 횟수 : {dice[idx]}')
    print(f'{total}번 중 {idx+1}이 나온 확률 : {dice[idx] / total}')
    print('-'*50)
````

````python
import random
import matplotlib.pyplot as plt

total = 10
dice = [0]*6

for i in range(total):
    num = random.randint(1,6)
    dice[num-1] += 1

for i in range(6):
    dice[i] /= total

plt.bar(range(1,7), dice)
plt.ylim(0,0.3)
plt.ylabel('prob')
plt.xlabel('dice number')
plt.title('rolling count 10')
plt.show()

plt.pie(dice,autopct='%2.2f%%',labels=range(1,7)) 
plt.show()
````

**주사위 2개 던져 합 확률 구하기**

````python
import random
import matplotlib.pyplot as plt

total = int(input('total: '))
dice = [0]*13 ##0 1 2 3 4 ~ 12

for i in range(total):
    num1 = random.randint(1,6)
    num2 = random.randint(1,6)
    dice[num1 + num2] += 1
    print(num1, num2, dice)

for i in range(13):
    dice[i] /= total
print(f'주사위 2개를 {total}회 던집니다!')
print(dice)
````
